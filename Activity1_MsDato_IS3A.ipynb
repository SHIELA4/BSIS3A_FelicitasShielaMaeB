```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/SHIELA4/BSIS3A_FelicitasShielaMaeB/refs/heads/main/TMDb_average_ratings.csv')
```

```python
print("original row count",len(df))
```

```python
print(df)
```

```python
df['vote_count'] = df['vote_count'].fillna(df['vote_count'].mean())
```

```python
print(df)
```

```python
df['vote_average'] = df['vote_average'].fillna(df['vote_average'].mean())
```

```python
print(df)
```

```python
df = df[df['vote_count'] >= 2000]
```

```python
print(df)
```

```python
df.to_csv('TMDb_cleaned.csv', index=False)
```
